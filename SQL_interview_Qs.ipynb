# Support for SQL server via PyODBC package

```python
 import pyodbc
 config = dict(server=   'localhost', # change this to your SQL Server hostname or IP address
               port=      1433,      # change this to your SQL Server port number [1433 is the default]
               database= 'AdventureWorks',
               username= 'tanya',
               password= 'Tanya1234')
 conn_str = ('SERVER={server},{port};'   +
             'DATABASE={database};'     +
             'UID={username};'          +
             'PWD={password}')
 conn = pyodbc.connect(
     r'DRIVER={ODBC Driver 13 for SQL Server};' +
     conn_str.format(**config)
     )
 cursor = conn.cursor()
 cursor.execute('SELECT * FROM dbo.AWBuildVersion;')
 for entry in cursor:
     print(entry)
 conn.close
```
